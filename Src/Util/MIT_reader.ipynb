```python
file_extensions = ['.atr', '.hea', '.dat', '.xws']
sample_folder = 'data/MIT/'

for extension in file_extensions:
    file_path = sample_folder + '100' + extension  # Change 'example' to the actual file name you have
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            print(f"Content of {file_path}:\n{content}\n{'-'*50}")
    except FileNotFoundError:
        print(f"{file_path} not found.")
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
```

```python
import numpy as np

dat_file_path = 'data/MIT/100.dat'

try:
    # Use NumPy to load binary data from the .dat file
    data = np.fromfile(dat_file_path, dtype=np.float32)  # Change dtype accordingly

    # Print the content of the data array
    print(f"Content of {dat_file_path}:\n{data}\n{'-' * 50}")

except FileNotFoundError:
    print(f"{dat_file_path} not found.")
except Exception as e:
    print(f"Error reading {dat_file_path}: {e}")
```

```python
file_extensions = ['.atr', '.hea', '.dat', '.xws']
sample_folder = 'data/MIT/'

for extension in file_extensions:
    file_path = sample_folder + '100' + extension  # Change 'example' to the actual file name you have
    try:
        with open(file_path, 'rb') as file:
            if extension == '.atr':
                content = file.read().hex()
            else:
                content = file.read().decode('utf-8')
            
            print(f"Content of {file_path}:\n{content}\n{'-'*50}")
    except FileNotFoundError:
        print(f"{file_path} not found.")
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
```
